```python
import requests
from bs4 import BeautifulSoup
```

```python
info = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

```python
resp = requests.get('http://www.melon.com/chart/index.htm', headers=info)
```

```python
html = resp.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
song_result = soup.select('.ellipsis.rank01 a')
singer_result = soup.select('.checkEllipsis')
```

```python
song_list = []
singer_list = []
rank = []
```

```python
for idx, song in enumerate(song_result):
    print(idx, song.text, singer_result[idx].text)
    rank.append(idx+1)
    song_list.append(song.text)
    singer_list.append(singer_result[idx].text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['순위'] = rank
```

```python
df['노래 제목'] = song_list
```

```python
df['가수'] = singer_list
```

```python
df
```

```python
df.to_excel('melon_Top_100.xlsx', index=False)
```
